```python
df= pd.read_csv("cancer patient data sets.csv")
d1=df.drop("Patient Id", axis=1)
dum= pd.get_dummies(d1, "Level", dtype= int)
d2= d1.drop("Level", axis=1)
bc= pd.concat([d2, dum], axis=1)
bc
```

```python
y = bc["Level_High"], bc["Level_Medium"], bc["Level_Low"]
x = d1.drop("Level", axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, shuffle=True, train_size=0.8, random_state=1889)
```
